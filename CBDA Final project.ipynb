```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
# ══════════════════════════════════════════════════════════════════════════════
# CREATE SQLITE DATA WAREHOUSE
# ══════════════════════════════════════════════════════════════════════════════

DB_PATH = 'healthcare_dw.db'
conn = sqlite3.connect(DB_PATH)
cursor = conn.cursor()

# ── Dimension: Patient ────────────────────────────────────────────────────────
cursor.execute('''
CREATE TABLE IF NOT EXISTS DIM_PATIENT (
    patient_key   INTEGER PRIMARY KEY AUTOINCREMENT,
    patient_id    TEXT UNIQUE NOT NULL,
    name          TEXT,
    dob           DATE,
    age           INTEGER,
    gender        TEXT CHECK(gender IN ('M','F','Other')),
    blood_group   TEXT,
    chronic_cond  TEXT,
    smoking       INTEGER DEFAULT 0,
    diabetes      INTEGER DEFAULT 0,
    hypertension  INTEGER DEFAULT 0,
    bmi           REAL,
    insurance     TEXT,
    created_at    TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
''')

# ── Dimension: Disease ────────────────────────────────────────────────────────
cursor.execute('''
CREATE TABLE IF NOT EXISTS DIM_DISEASE (
    disease_key    INTEGER PRIMARY KEY AUTOINCREMENT,
    icd10_code     TEXT UNIQUE NOT NULL,
    disease_name   TEXT NOT NULL,
    category       TEXT,
    sub_category   TEXT,
    transmission   TEXT,
    severity_level TEXT CHECK(severity_level IN ('Low','Moderate','High','Critical')),
    notifiable     INTEGER DEFAULT 0,
    endemic_region TEXT
);
''')

# ── Dimension: Location ───────────────────────────────────────────────────────
cursor.execute('''
CREATE TABLE IF NOT EXISTS DIM_LOCATION (
    location_key   INTEGER PRIMARY KEY AUTOINCREMENT,
    zone_id        TEXT,
    zone_name      TEXT,
    district       TEXT,
    state          TEXT,
    latitude       REAL,
    longitude      REAL,
    population     INTEGER,
    density        TEXT,
    water_quality  TEXT,
    sanitation_idx REAL,
    hospital_count INTEGER
);
''')

# ── Dimension: Time ──────────────────────────────────────────────────────────
cursor.execute('''
CREATE TABLE IF NOT EXISTS DIM_TIME (
    time_key    INTEGER PRIMARY KEY AUTOINCREMENT,
    date_val    DATE UNIQUE NOT NULL,
    day         INTEGER,
    week        INTEGER,
    month       INTEGER,
    quarter     INTEGER,
    year        INTEGER,
    season      TEXT,
    is_monsoon  INTEGER DEFAULT 0,
    is_holiday  INTEGER DEFAULT 0
);
''')

# ── Dimension: Lab ────────────────────────────────────────────────────────────
cursor.execute('''
CREATE TABLE IF NOT EXISTS DIM_LAB (
    lab_key       INTEGER PRIMARY KEY AUTOINCREMENT,
    lab_id        TEXT UNIQUE,
    lab_name      TEXT,
    test_type     TEXT,
    result_value  REAL,
    unit          TEXT,
    normal_range  TEXT,
    flag          TEXT CHECK(flag IN ('Normal','Abnormal','Critical')),
    pathogen_det  TEXT
);
''')

# ── Dimension: Pharmacy ──────────────────────────────────────────────────────
cursor.execute('''
CREATE TABLE IF NOT EXISTS DIM_PHARMACY (
    pharmacy_key  INTEGER PRIMARY KEY AUTOINCREMENT,
    rx_id         TEXT UNIQUE,
    drug_name     TEXT,
    drug_class    TEXT,
    dosage        TEXT,
    dispensed     INTEGER DEFAULT 0,
    antibiotic    INTEGER DEFAULT 0,
    antiviral     INTEGER DEFAULT 0
);
''')

# ── Fact Table: Case ─────────────────────────────────────────────────────────
cursor.execute('''
CREATE TABLE IF NOT EXISTS FACT_CASE (
    case_key          INTEGER PRIMARY KEY AUTOINCREMENT,
    case_id           TEXT UNIQUE NOT NULL,
    patient_key       INTEGER REFERENCES DIM_PATIENT(patient_key),
    disease_key       INTEGER REFERENCES DIM_DISEASE(disease_key),
    location_key      INTEGER REFERENCES DIM_LOCATION(location_key),
    time_key          INTEGER REFERENCES DIM_TIME(time_key),
    lab_key           INTEGER REFERENCES DIM_LAB(lab_key),
    pharmacy_key      INTEGER REFERENCES DIM_PHARMACY(pharmacy_key),
    -- Measures
    severity_score    REAL,
    hospitalized      INTEGER DEFAULT 0,
    icu_admitted      INTEGER DEFAULT 0,
    days_to_recovery  INTEGER,
    mortality         INTEGER DEFAULT 0,
    contacts_traced   INTEGER DEFAULT 0,
    symptom_onset     DATE,
    confirmed_date    DATE,
    outbreak_flag     INTEGER DEFAULT 0,
    cluster_id        INTEGER DEFAULT -1,
    risk_score        REAL
);
''')

# ── Fact Table: Outbreak Alert Log ───────────────────────────────────────────
cursor.execute('''
CREATE TABLE IF NOT EXISTS FACT_OUTBREAK_ALERT (
    alert_id       INTEGER PRIMARY KEY AUTOINCREMENT,
    alert_date     DATE,
    disease_name   TEXT,
    zone_name      TEXT,
    case_count     INTEGER,
    threshold      INTEGER,
    alert_level    TEXT CHECK(alert_level IN ('Watch','Warning','Emergency')),
    triggered_by   TEXT,
    resolved       INTEGER DEFAULT 0
);
''')

conn.commit()
print('✅ Data Warehouse schema created: 6 tables (4 dims + 2 facts)')
print(f'   📂 Database: {DB_PATH}')

# Print table list
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", conn)
print('\n📊 Tables in Warehouse:')
for t in tables['name']:
    print(f'   → {t}')
```

```
# ══════════════════════════════════════════════════════════════════════════════
# ERD VISUALIZATION (Matplotlib)
# ══════════════════════════════════════════════════════════════════════════════

fig, ax = plt.subplots(figsize=(16, 11))
ax.set_xlim(0, 16)
ax.set_ylim(0, 11)
ax.axis('off')
ax.set_facecolor('#0d1117')
fig.patch.set_facecolor('#0d1117')

# Color palette
fact_color   = '#1f6feb'
dim_color    = '#388bfd'
alert_color  = '#da3633'
line_color   = '#58a6ff'
text_main    = '#e6edf3'
text_sub     = '#8b949e'

def draw_table(ax, x, y, title, fields, color, width=2.8, row_h=0.28):
    h = row_h * (len(fields) + 1) + 0.1
    # Header
    rect_h = plt.Rectangle((x, y), width, row_h + 0.05, color=color, zorder=3)
    ax.add_patch(rect_h)
    ax.text(x + width/2, y + row_h/2 + 0.02, title, ha='center', va='center',
            fontsize=7.5, fontweight='bold', color='white', zorder=4)
    # Body
    rect_b = plt.Rectangle((x, y - row_h * len(fields)), width,
                            row_h * len(fields), color='#161b22', zorder=2)
    border = plt.Rectangle((x, y - row_h * len(fields)), width,
                            row_h * len(fields) + row_h + 0.05,
                            fill=False, edgecolor=color, linewidth=1.2, zorder=5)
    ax.add_patch(rect_b)
    ax.add_patch(border)
    for i, (fname, ftype, pk) in enumerate(fields):
        fy = y - row_h * (i + 0.6)
        icon = '🔑 ' if pk == 'PK' else ('🔗 ' if pk == 'FK' else '   ')
        ax.text(x + 0.08, fy, f'{icon}{fname}', ha='left', va='center',
                fontsize=5.8, color=text_main if pk else text_sub, zorder=6)
        ax.text(x + width - 0.08, fy, ftype, ha='right', va='center',
                fontsize=5, color='#3fb950', zorder=6)
    return (x + width/2, y), (x + width/2, y - row_h * len(fields))

# ── Tables ────────────────────────────────────────────────────────────────────
fact_fields = [
    ('case_key','INT','PK'), ('case_id','TEXT',''), ('patient_key','INT','FK'),
    ('disease_key','INT','FK'), ('location_key','INT','FK'), ('time_key','INT','FK'),
    ('lab_key','INT','FK'), ('pharmacy_key','INT','FK'), ('severity_score','REAL',''),
    ('hospitalized','INT',''), ('mortality','INT',''), ('outbreak_flag','INT',''),
    ('risk_score','REAL',''),
]

patient_fields = [('patient_key','INT','PK'),('patient_id','TEXT',''),('age','INT',''),
                  ('gender','TEXT',''),('blood_group','TEXT',''),('chronic_cond','TEXT',''),
                  ('smoking','INT',''),('diabetes','INT',''),('hypertension','INT',''),('bmi','REAL','')]

disease_fields = [('disease_key','INT','PK'),('icd10_code','TEXT',''),('disease_name','TEXT',''),
                  ('category','TEXT',''),('severity_level','TEXT',''),('transmission','TEXT',''),
                  ('notifiable','INT','')]

location_fields = [('location_key','INT','PK'),('zone_id','TEXT',''),('zone_name','TEXT',''),
                   ('latitude','REAL',''),('longitude','REAL',''),('population','INT',''),
                   ('sanitation_idx','REAL','')]

time_fields = [('time_key','INT','PK'),('date_val','DATE',''),('month','INT',''),
               ('quarter','INT',''),('year','INT',''),('season','TEXT',''),('is_monsoon','INT','')]

lab_fields = [('lab_key','INT','PK'),('lab_id','TEXT',''),('test_type','TEXT',''),
              ('result_value','REAL',''),('flag','TEXT',''),('pathogen_det','TEXT','')]

pharma_fields = [('pharmacy_key','INT','PK'),('rx_id','TEXT',''),('drug_name','TEXT',''),
                 ('drug_class','TEXT',''),('antibiotic','INT',''),('antiviral','INT','')]

alert_fields = [('alert_id','INT','PK'),('alert_date','DATE',''),('disease_name','TEXT',''),
                ('zone_name','TEXT',''),('case_count','INT',''),('alert_level','TEXT','')]

# Draw
draw_table(ax, 5.6, 7.8, 'FACT_CASE (Central Fact)', fact_fields, fact_color, width=3.0)
draw_table(ax, 0.4, 10.2, 'DIM_PATIENT', patient_fields, dim_color)
draw_table(ax, 8.8, 10.2, 'DIM_DISEASE', disease_fields, dim_color)
draw_table(ax, 0.4, 6.0,  'DIM_LOCATION', location_fields, dim_color)
draw_table(ax, 8.8, 6.0,  'DIM_TIME', time_fields, dim_color)
draw_table(ax, 0.4, 2.2,  'DIM_LAB', lab_fields, dim_color)
draw_table(ax, 8.8, 2.2,  'DIM_PHARMACY', pharma_fields, dim_color)
draw_table(ax, 5.6, 2.5,  'FACT_OUTBREAK_ALERT', alert_fields, alert_color, width=3.0)

# Connector lines
conns = [
    (3.2, 8.8,  5.6, 7.8), (11.6, 8.8, 8.6, 7.8),
    (3.2, 4.5,  5.6, 5.2), (11.6, 4.5, 8.6, 5.2),
    (3.2, 1.0,  5.6, 3.5), (11.6, 1.0, 8.6, 3.5),
    (7.1, 4.2,  7.1, 2.5),
]
for (x1,y1,x2,y2) in conns:
    ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
                arrowprops=dict(arrowstyle='->', color=line_color, lw=1.5,
                                connectionstyle='arc3,rad=0.0'))

ax.text(8, 10.8, 'HEALTHCARE DATA WAREHOUSE — STAR SCHEMA ERD',
        ha='center', va='center', fontsize=13, fontweight='bold',
        color=text_main, transform=ax.transData)

# Legend
legend_elements = [
    mpatches.Patch(color=fact_color,  label='Fact Table'),
    mpatches.Patch(color=dim_color,   label='Dimension Table'),
    mpatches.Patch(color=alert_color, label='Alert Log Table'),
]
ax.legend(handles=legend_elements, loc='lower left', facecolor='#161b22',
          edgecolor='#30363d', labelcolor=text_main, fontsize=8)

plt.tight_layout()
plt.savefig('erd_schema.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()
print('✅ ERD saved as erd_schema.png')
```

```
# ══════════════════════════════════════════════════════════════════════════════
# ETL STEP 1: EXTRACT — Generate source-system records
# ══════════════════════════════════════════════════════════════════════════════

def extract_hospital_mis(n):
    """Simulate Hospital MIS patient extract"""
    blood_groups = ['A+','A-','B+','B-','AB+','AB-','O+','O-']
    chronic_list = ['None','Diabetes','Hypertension','Asthma','CKD','Heart Disease']
    rows = []
    for i in range(n):
        age   = int(np.random.choice(
            range(1, 85),
            p=np.array([1]*84)/84
        ))
        bmi   = round(np.random.normal(24.5, 4.5), 1)
        diab  = int(np.random.random() < (0.25 if age > 40 else 0.05))
        hyper = int(np.random.random() < (0.30 if age > 45 else 0.08))
        smoke = int(np.random.random() < 0.18)
        rows.append({
            'patient_id':   f'P{i+1:05d}',
            'name':         fake.name(),
            'dob':          str(datetime.now().date() - timedelta(days=age*365)),
            'age':          age,
            'gender':       np.random.choice(['M','F'], p=[0.52, 0.48]),
            'blood_group':  random.choice(blood_groups),
            'chronic_cond': random.choice(chronic_list),
            'smoking':      smoke,
            'diabetes':     diab,
            'hypertension': hyper,
            'bmi':          max(14.0, min(45.0, bmi)),
            'insurance':    np.random.choice(['CGHS','Private','None','Aarogya Setu'], p=[0.15,0.3,0.4,0.15])
        })
    return pd.DataFrame(rows)


def extract_lab_system(n):
    """Simulate Lab Information System extract"""
    tests  = ['CBC','Dengue NS1','Widal','Blood Culture','PCR-Covid','Malaria Smear',
               'Sputum AFB','LFT','KFT','Chest X-Ray','Urine Culture','ELISA']
    rows = []
    for i in range(n):
        val  = round(np.random.uniform(0.1, 500.0), 2)
        flag = np.random.choice(['Normal','Abnormal','Critical'], p=[0.55, 0.35, 0.10])
        rows.append({
            'lab_id':       f'L{i+1:05d}',
            'lab_name':     fake.company() + ' Labs',
            'test_type':    random.choice(tests),
            'result_value': val,
            'unit':         random.choice(['mg/dL','cells/uL','IU/L','g/dL','%']),
            'normal_range': '70-110',
            'flag':         flag,
            'pathogen_det': random.choice(['None','Dengue Virus','Salmonella',
                                           'Plasmodium','M.tuberculosis','None','None'])
        })
    return pd.DataFrame(rows)


def extract_pharmacy(n):
    """Simulate Pharmacy Management System extract"""
    drugs = [
        ('Paracetamol','Analgesic',0,0), ('Amoxicillin','Penicillin Antibiotic',1,0),
        ('Ciprofloxacin','Fluoroquinolone',1,0), ('Oseltamivir','Antiviral',0,1),
        ('Chloroquine','Antimalarial',0,0), ('ORS','Rehydration',0,0),
        ('Artemether','Antimalarial',0,0), ('Rifampicin','Anti-TB',1,0),
        ('Doxycycline','Tetracycline',1,0), ('Ibuprofen','NSAID',0,0),
    ]
    rows = []
    for i in range(n):
        drug = random.choice(drugs)
        rows.append({
            'rx_id':      f'RX{i+1:05d}',
            'drug_name':  drug[0],
            'drug_class': drug[1],
            'dosage':     random.choice(['250mg','500mg','1g','10mg/kg']),
            'dispensed':  int(np.random.random() < 0.88),
            'antibiotic': drug[2],
            'antiviral':  drug[3]
        })
    return pd.DataFrame(rows)


# Run extractions
df_patients = extract_hospital_mis(N_PATIENTS)
df_labs     = extract_lab_system(N_CASES)
df_pharma   = extract_pharmacy(N_CASES)

print('✅ ETL — EXTRACT complete:')
print(f'   📋 Patients:    {len(df_patients):,} rows')
print(f'   🧪 Lab records: {len(df_labs):,} rows')
print(f'   💊 Pharmacy:    {len(df_pharma):,} rows')
print('\n📄 Sample Patient Record:')
display(df_patients.head(3))
```

```
# ══════════════════════════════════════════════════════════════════════════════
# ETL STEP 2: TRANSFORM — Clean, validate, enrich
# ══════════════════════════════════════════════════════════════════════════════

def transform_patients(df):
    """Transform & validate patient data"""
    df = df.copy()
    # Fix BMI outliers
    df['bmi'] = df['bmi'].clip(13, 50)
    # Age bins
    df['age_group'] = pd.cut(df['age'],
                              bins=[0,5,18,35,60,100],
                              labels=['Child','Teen','Adult','Middle-Aged','Senior'])
    # Risk factor composite
    df['risk_factors'] = df['smoking'] + df['diabetes'] + df['hypertension']
    # Impute missing
    df['gender'].fillna('M', inplace=True)
    return df

def transform_labs(df):
    """Transform lab data — derive severity"""
    df = df.copy()
    df['severity_score'] = df['flag'].map({'Normal': 0, 'Abnormal': 1, 'Critical': 2})
    df['result_value']   = df['result_value'].clip(0, 1000)
    return df

# Apply transforms
df_patients_t = transform_patients(df_patients)
df_labs_t     = transform_labs(df_labs)

print('✅ ETL — TRANSFORM complete')
print('\n📊 Patient age distribution after transform:')
print(df_patients_t['age_group'].value_counts())
print('\n🧪 Lab flag distribution:')
print(df_labs_t['flag'].value_counts())
```

```
# ══════════════════════════════════════════════════════════════════════════════
# ETL STEP 3: LOAD — Populate all warehouse dimensions and facts
# ══════════════════════════════════════════════════════════════════════════════

# ── Load DIM_DISEASE ──────────────────────────────────────────────────────────
for d in DISEASES:
    cursor.execute('''
        INSERT OR IGNORE INTO DIM_DISEASE
        (icd10_code, disease_name, category, sub_category, transmission, severity_level, notifiable)
        VALUES (?,?,?,?,?,?,?)''', d)

# ── Load DIM_LOCATION ─────────────────────────────────────────────────────────
for z in ZONES:
    cursor.execute('''
        INSERT OR IGNORE INTO DIM_LOCATION
        (zone_id,zone_name,district,state,latitude,longitude,population,density,water_quality,sanitation_idx,hospital_count)
        VALUES (?,?,?,?,?,?,?,?,?,?,?)''', z)

# ── Load DIM_TIME ─────────────────────────────────────────────────────────────
all_dates = pd.date_range(START_DATE, END_DATE, freq='D')
for d in all_dates:
    month   = d.month
    season  = ('Summer' if month in [3,4,5] else
                'Monsoon' if month in [6,7,8,9] else
                'Post-Monsoon' if month in [10,11] else 'Winter')
    is_monsoon = int(month in [6,7,8,9])
    cursor.execute('''
        INSERT OR IGNORE INTO DIM_TIME
        (date_val,day,week,month,quarter,year,season,is_monsoon)
        VALUES (?,?,?,?,?,?,?,?)''',
        (str(d.date()), d.day, d.isocalendar()[1], month,
         d.quarter, d.year, season, is_monsoon))

# ── Load DIM_PATIENT ──────────────────────────────────────────────────────────
df_patients_t[['patient_id','name','dob','age','gender','blood_group',
               'chronic_cond','smoking','diabetes','hypertension',
               'bmi','insurance']].to_sql(
    'DIM_PATIENT', conn, if_exists='append', index=False
)

# ── Load DIM_LAB ──────────────────────────────────────────────────────────────
df_labs_t[['lab_id','lab_name','test_type','result_value',
           'unit','normal_range','flag','pathogen_det']].to_sql(
    'DIM_LAB', conn, if_exists='append', index=False
)

# ── Load DIM_PHARMACY ─────────────────────────────────────────────────────────
df_pharma[['rx_id','drug_name','drug_class','dosage',
           'dispensed','antibiotic','antiviral']].to_sql(
    'DIM_PHARMACY', conn, if_exists='append', index=False
)

# ── Fetch keys ────────────────────────────────────────────────────────────────
patient_keys  = pd.read_sql('SELECT patient_key, patient_id FROM DIM_PATIENT', conn)
disease_keys  = pd.read_sql('SELECT disease_key, disease_name FROM DIM_DISEASE', conn)
location_keys = pd.read_sql('SELECT location_key, zone_id FROM DIM_LOCATION', conn)
time_keys     = pd.read_sql('SELECT time_key, date_val FROM DIM_TIME', conn)
lab_keys      = pd.read_sql('SELECT lab_key, lab_id FROM DIM_LAB', conn)
pharma_keys   = pd.read_sql('SELECT pharmacy_key, rx_id FROM DIM_PHARMACY', conn)

conn.commit()
print('✅ All dimension tables loaded')
print(f'   Patients: {len(patient_keys):,} | Diseases: {len(disease_keys)} | Locations: {len(location_keys)}')
print(f'   Time entries: {len(time_keys):,} | Lab records: {len(lab_keys):,} | Pharmacy: {len(pharma_keys):,}')
```

```
# ── Generate & Load FACT_CASE ─────────────────────────────────────────────────

# Disease-specific seasonal weights (higher in certain months)
disease_month_weight = {
    'Dengue Fever':        {6:3, 7:5, 8:6, 9:5, 10:3},  # monsoon peak
    'Malaria':             {6:3, 7:4, 8:5, 9:4, 10:2},
    'Typhoid Fever':       {4:2, 5:3, 6:3, 7:2},
    'Cholera':             {5:2, 6:4, 7:4},
    'Influenza':           {11:3, 12:4, 1:5, 2:4, 3:2},  # winter peak
    'Pneumonia':           {11:3, 12:4, 1:4, 2:3},
    'Chikungunya':         {7:3, 8:4, 9:3},
    'Leptospirosis':       {7:3, 8:4, 9:3},
}

def get_disease_weight(dname, month):
    return disease_month_weight.get(dname, {}).get(month, 1)

print('⏳ Generating fact records (this may take 20-30s)...')

fact_records = []
date_list = list(pd.date_range(START_DATE, END_DATE, freq='D'))

for i in range(N_CASES):
    # Sample date with monsoon bias
    dt      = random.choice(date_list)
    p_row   = patient_keys.sample(1).iloc[0]
    d_row   = disease_keys.sample(1).iloc[0]
    l_row   = location_keys.sample(1).iloc[0]
    lab_row = lab_keys.iloc[i % len(lab_keys)]
    ph_row  = pharma_keys.iloc[i % len(pharma_keys)]

    # Time key lookup
    tk_row  = time_keys[time_keys['date_val'] == str(dt.date())]
    if tk_row.empty:
        continue
    tk = tk_row.iloc[0]['time_key']

    sev   = round(np.random.beta(2, 5) * 10, 2)
    hosp  = int(sev > 6)
    icu   = int(sev > 8.5)
    mort  = int(sev > 9.2 and np.random.random() < 0.3)
    recov = int(np.random.exponential(7)) + 1
    contacts = int(np.random.poisson(3))
    risk  = round(np.clip(sev/10 + np.random.normal(0, 0.1), 0, 1), 3)
    outbk = int(risk > 0.75)

    fact_records.append((
        f'CASE{i+1:06d}',
        int(p_row['patient_key']),
        int(d_row['disease_key']),
        int(l_row['location_key']),
        int(tk),
        int(lab_row['lab_key']),
        int(ph_row['pharmacy_key']),
        sev, hosp, icu, recov, mort,
        contacts,
        str((dt - timedelta(days=random.randint(1,5))).date()),
        str(dt.date()),
        outbk, -1, risk
    ))

cursor.executemany('''
    INSERT OR IGNORE INTO FACT_CASE
    (case_id, patient_key, disease_key, location_key, time_key,
     lab_key, pharmacy_key, severity_score, hospitalized, icu_admitted,
     days_to_recovery, mortality, contacts_traced, symptom_onset,
     confirmed_date, outbreak_flag, cluster_id, risk_score)
    VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)''',
    fact_records
)

conn.commit()
loaded = cursor.execute('SELECT COUNT(*) FROM FACT_CASE').fetchone()[0]
print(f'✅ ETL — LOAD complete: {loaded:,} fact records loaded into FACT_CASE')
```

```
# ══════════════════════════════════════════════════════════════════════════════
# FEATURE ENGINEERING
# ══════════════════════════════════════════════════════════════════════════════

ml_df = df.copy()

# ── Encode categoricals ──────────────────────────────────────────────────────
le = LabelEncoder()
ml_df['gender_enc']       = le.fit_transform(ml_df['gender'])
ml_df['season_enc']       = le.fit_transform(ml_df['season'])
ml_df['water_qual_enc']   = le.fit_transform(ml_df['water_quality'])
ml_df['lab_flag_enc']     = le.fit_transform(ml_df['lab_flag'])
ml_df['sev_level_enc']    = le.fit_transform(ml_df['severity_level'])
ml_df['transmission_enc'] = le.fit_transform(ml_df['transmission'])

# ── Feature list ─────────────────────────────────────────────────────────────
FEATURES = [
    'age', 'gender_enc', 'smoking', 'diabetes', 'hypertension', 'bmi',
    'severity_score', 'month', 'is_monsoon', 'quarter',
    'sanitation_idx', 'water_qual_enc', 'lab_flag_enc',
    'sev_level_enc', 'transmission_enc', 'notifiable',
    'antibiotic', 'antiviral', 'contacts_traced', 'risk_score'
]

TARGET = 'hospitalized'

X = ml_df[FEATURES].fillna(0)
y = ml_df[TARGET]

# ── SMOTE for class imbalance ─────────────────────────────────────────────────
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42, stratify=y_res
)

scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)

print(f'✅ Feature engineering complete')
print(f'   Features: {len(FEATURES)} | Train: {len(X_train):,} | Test: {len(X_test):,}')
print(f'   Class balance after SMOTE: {pd.Series(y_res).value_counts().to_dict()}')
```

```
# ══════════════════════════════════════════════════════════════════════════════
# TRAIN & EVALUATE 3 MODELS
# ══════════════════════════════════════════════════════════════════════════════

models = {
    'Logistic Regression': LogisticRegression(max_iter=500, random_state=42),
    'Random Forest':       RandomForestClassifier(n_estimators=200, max_depth=12,
                                                   min_samples_leaf=5, random_state=42, n_jobs=-1),
    'XGBoost':             xgb.XGBClassifier(n_estimators=200, max_depth=6, learning_rate=0.1,
                                              subsample=0.8, colsample_bytree=0.8,
                                              use_label_encoder=False, eval_metric='logloss',
                                              random_state=42, verbosity=0)
}

results = {}
trained_models = {}

for name, model in models.items():
    print(f'\n🔄 Training {name}...')
    Xtr = X_train_sc if name == 'Logistic Regression' else X_train
    Xts = X_test_sc  if name == 'Logistic Regression' else X_test

    model.fit(Xtr, y_train)
    y_pred  = model.predict(Xts)
    y_prob  = model.predict_proba(Xts)[:, 1]

    acc     = accuracy_score(y_test, y_pred)
    auc     = roc_auc_score(y_test, y_prob)
    cv_sc   = cross_val_score(model, Xtr, y_train, cv=5, scoring='accuracy')

    results[name] = {
        'accuracy': acc, 'auc': auc,
        'cv_mean': cv_sc.mean(), 'cv_std': cv_sc.std(),
        'y_pred': y_pred, 'y_prob': y_prob
    }
    trained_models[name] = model

    print(f'   ✅ Accuracy: {acc:.4f} | AUC: {auc:.4f} | CV: {cv_sc.mean():.4f} ± {cv_sc.std():.4f}')

print('\n' + '='*55)
print('📊 MODEL COMPARISON SUMMARY')
print('='*55)
print(f'{"Model":<22} {"Accuracy":>10} {"AUC":>8} {"CV Mean":>10}')
print('-'*55)
for name, r in results.items():
    tag = ' ← BEST' if r['accuracy'] == max(v['accuracy'] for v in results.values()) else ''
    print(f'{name:<22} {r["accuracy"]:>10.4f} {r["auc"]:>8.4f} {r["cv_mean"]:>10.4f}{tag}')
```

```
# ══════════════════════════════════════════════════════════════════════════════
# SPATIAL ANALYSIS 1 — DBSCAN Clustering
# ══════════════════════════════════════════════════════════════════════════════

# Add jitter to lat/lon for spatial spread
df['lat_jitter'] = df['latitude']  + np.random.normal(0, 0.08, len(df))
df['lon_jitter'] = df['longitude'] + np.random.normal(0, 0.08, len(df))

coords = df[['lat_jitter','lon_jitter']].values

# DBSCAN spatial clustering (epsilon=0.05 degrees ≈ 5km)
dbscan = DBSCAN(eps=0.05, min_samples=15, metric='haversine')
coords_rad = np.radians(coords)
labels = dbscan.fit_predict(coords_rad)

df['cluster_id'] = labels
n_clusters  = len(set(labels)) - (1 if -1 in labels else 0)
n_noise     = sum(labels == -1)

print(f'✅ DBSCAN Spatial Clustering:')
print(f'   Clusters found:    {n_clusters}')
print(f'   Noise points:      {n_noise:,}')
print(f'   Clustered points:  {sum(labels != -1):,}')

cluster_stats = df[df['cluster_id'] != -1].groupby('cluster_id').agg(
    count=('case_id','count'),
    lat=('lat_jitter','mean'),
    lon=('lon_jitter','mean'),
    hosp_rate=('hospitalized','mean'),
    severity=('severity_score','mean')
).reset_index()

print('\n📍 Top 5 Clusters by Case Count:')
display(cluster_stats.sort_values('count', ascending=False).head(5).round(3))
```

```
# ══════════════════════════════════════════════════════════════════════════════
# OUTBREAK DETECTION — Multiple Statistical Methods
# ══════════════════════════════════════════════════════════════════════════════

# Disease-specific outbreak thresholds (weekly cases per zone)
OUTBREAK_THRESHOLDS = {
    'Dengue Fever':        {'Watch': 5,  'Warning': 15, 'Emergency': 30},
    'Typhoid Fever':       {'Watch': 3,  'Warning': 10, 'Emergency': 20},
    'Pneumonia':           {'Watch': 8,  'Warning': 20, 'Emergency': 40},
    'Gastroenteritis':     {'Watch': 10, 'Warning': 25, 'Emergency': 50},
    'Malaria':             {'Watch': 4,  'Warning': 12, 'Emergency': 25},
    'Tuberculosis':        {'Watch': 2,  'Warning': 6,  'Emergency': 12},
    'Acute Resp Infection':{'Watch': 15, 'Warning': 40, 'Emergency': 80},
    'Leptospirosis':       {'Watch': 2,  'Warning': 5,  'Emergency': 10},
    'HIV/AIDS':            {'Watch': 1,  'Warning': 3,  'Emergency': 6},
    'Rabies':              {'Watch': 1,  'Warning': 2,  'Emergency': 4},
    'Influenza':           {'Watch': 10, 'Warning': 30, 'Emergency': 60},
    'Chikungunya':         {'Watch': 4,  'Warning': 12, 'Emergency': 24},
}

def classify_alert(count, thresholds):
    if count >= thresholds['Emergency']: return 'Emergency'
    if count >= thresholds['Warning']:   return 'Warning'
    if count >= thresholds['Watch']:     return 'Watch'
    return None


# ── Method 1: Threshold-based ─────────────────────────────────────────────────
df['week'] = df['confirmed_date'].dt.isocalendar().week
df['year'] = df['confirmed_date'].dt.year

weekly_zone = df.groupby(['year','week','disease_name','zone_name']).size().reset_index(name='case_count')

alerts_threshold = []
for _, row in weekly_zone.iterrows():
    thresh = OUTBREAK_THRESHOLDS.get(row['disease_name'], {'Watch':5,'Warning':15,'Emergency':30})
    level  = classify_alert(row['case_count'], thresh)
    if level:
        alerts_threshold.append({
            'year': row['year'], 'week': row['week'],
            'disease_name': row['disease_name'],
            'zone_name': row['zone_name'],
            'case_count': row['case_count'],
            'threshold': thresh['Warning'],
            'alert_level': level,
            'method': 'Threshold'
        })

df_alerts_t = pd.DataFrame(alerts_threshold)


# ── Method 2: Z-score anomaly detection ─────────────────────────────────────
alerts_zscore = []
for disease in df['disease_name'].unique():
    for zone in df['zone_name'].unique():
        sub = weekly_zone[(weekly_zone['disease_name']==disease) &
                          (weekly_zone['zone_name']==zone)].copy()
        if len(sub) < 4:
            continue
        sub['z_score'] = stats.zscore(sub['case_count'])
        anomalies = sub[sub['z_score'] > 2.0]
        for _, row in anomalies.iterrows():
            level = 'Emergency' if row['z_score'] > 3.5 else 'Warning' if row['z_score'] > 2.5 else 'Watch'
            alerts_zscore.append({
                'year': row['year'], 'week': row['week'],
                'disease_name': disease, 'zone_name': zone,
                'case_count': row['case_count'],
                'z_score': round(row['z_score'], 2),
                'alert_level': level, 'method': 'Z-Score'
            })

df_alerts_z = pd.DataFrame(alerts_zscore)


# ── Method 3: Cumulative sum (CUSUM) ─────────────────────────────────────────
def cusum_detect(series, k=0.5, h=4):
    """CUSUM change detection"""
    mu  = series.mean()
    sig = series.std() + 1e-6
    S_pos, S_neg = 0, 0
    alarms = []
    for i, x in enumerate(series):
        z = (x - mu) / sig
        S_pos = max(0, S_pos + z - k)
        S_neg = max(0, S_neg - z - k)
        if S_pos > h or S_neg > h:
            alarms.append(i)
            S_pos, S_neg = 0, 0
    return alarms

alerts_cusum = []
for disease in df['disease_name'].unique():
    sub = df[df['disease_name']==disease].groupby(
        df['confirmed_date'].dt.to_period('W')
    ).size()
    if len(sub) < 8:
        continue
    alarm_idxs = cusum_detect(sub.values)
    for idx in alarm_idxs:
        alerts_cusum.append({
            'disease_name': disease,
            'week_period': str(sub.index[idx]),
            'case_count': int(sub.values[idx]),
            'alert_level': 'Warning',
            'method': 'CUSUM'
        })

df_alerts_c = pd.DataFrame(alerts_cusum)


# ── Save alerts to warehouse ──────────────────────────────────────────────────
for _, row in df_alerts_t.iterrows():
    cursor.execute('''
        INSERT INTO FACT_OUTBREAK_ALERT
        (alert_date, disease_name, zone_name, case_count, threshold, alert_level, triggered_by)
        VALUES (?,?,?,?,?,?,?)''',
        (f"{row['year']}-W{row['week']:02d}",
         row['disease_name'], row['zone_name'],
         int(row['case_count']), int(row['threshold']),
         row['alert_level'], 'Threshold-Based')
    )
conn.commit()

print('✅ Outbreak detection complete:')
print(f'   📊 Threshold alerts:  {len(df_alerts_t):,}')
print(f'   📊 Z-score alerts:    {len(df_alerts_z):,}')
print(f'   📊 CUSUM alerts:      {len(df_alerts_c):,}')

print('\n🚨 Alert Level Summary (Threshold Method):')
if not df_alerts_t.empty:
    display(df_alerts_t['alert_level'].value_counts().to_frame('Count'))

print('\n🔴 Top Emergency Alerts:')
if not df_alerts_t.empty:
    emg = df_alerts_t[df_alerts_t['alert_level']=='Emergency'].sort_values('case_count', ascending=False)
    display(emg[['year','week','disease_name','zone_name','case_count','alert_level']].head(10))
```

```
# ══════════════════════════════════════════════════════════════════════════════
# TIME SERIES: ARIMA FORECASTING FOR DENGUE
# ══════════════════════════════════════════════════════════════════════════════

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Prepare weekly Dengue time series
dengue_ts = df[df['disease_name']=='Dengue Fever'].copy()
dengue_ts = dengue_ts.set_index('confirmed_date').resample('W').size()
dengue_ts.name = 'cases'

# Remove zero-only tail
dengue_ts = dengue_ts[dengue_ts.index >= '2021-01-01']

# ── Seasonal Decomposition ───────────────────────────────────────────────────
fig, axes = plt.subplots(4, 1, figsize=(16, 12))
fig.patch.set_facecolor('#0f172a')
for ax in axes:
    ax.set_facecolor('#1e293b')
    for sp in ax.spines.values():
        sp.set_edgecolor('#334155')

if len(dengue_ts) >= 52:
    decomp = seasonal_decompose(dengue_ts, model='additive', period=52)
    for ax, (label, data) in zip(axes, [
        ('Observed', dengue_ts),
        ('Trend', decomp.trend),
        ('Seasonal', decomp.seasonal),
        ('Residual', decomp.resid)
    ]):
        ax.plot(data.index, data.values, color='#60a5fa', linewidth=1.5)
        ax.set_ylabel(label, color='white', fontsize=10)
        ax.tick_params(colors='#94a3b8', labelsize=8)
        ax.fill_between(data.index, 0, data.values,
                         alpha=0.15, color='#3b82f6')
else:
    for ax, label in zip(axes, ['Observed','Trend','Seasonal','Residual']):
        ax.plot(dengue_ts.index, dengue_ts.values, color='#60a5fa', linewidth=1.5)
        ax.set_ylabel(label, color='white', fontsize=10)
        ax.tick_params(colors='#94a3b8', labelsize=8)

axes[0].set_title('Dengue Fever — Seasonal Decomposition (Weekly Cases)', color='white',
                   fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('time_series_decomp.png', dpi=130, bbox_inches='tight', facecolor='#0f172a')
plt.show()

# ── ARIMA Model ───────────────────────────────────────────────────────────────
train_ts = dengue_ts[:-12]
test_ts  = dengue_ts[-12:]

arima = ARIMA(train_ts, order=(2, 1, 2), seasonal_order=(1, 1, 1, 52))
arima_fit = arima.fit()

forecast    = arima_fit.forecast(steps=24)  # 24-week ahead
conf_int    = arima_fit.get_forecast(steps=24).conf_int()

# Plot forecast
fig, ax = plt.subplots(figsize=(16, 6))
fig.patch.set_facecolor('#0f172a')
ax.set_facecolor('#1e293b')

ax.plot(train_ts.index, train_ts.values, color='#60a5fa', linewidth=1.5, label='Historical')
ax.plot(test_ts.index,  test_ts.values,  color='#34d399', linewidth=1.5, label='Actual (test)')

forecast_idx = pd.date_range(dengue_ts.index[-1], periods=25, freq='W')[1:]
ax.plot(forecast_idx, forecast.values, color='#f97316', linewidth=2,
        linestyle='--', label='ARIMA Forecast')
ax.fill_between(forecast_idx,
                conf_int.iloc[:, 0], conf_int.iloc[:, 1],
                alpha=0.2, color='#f97316', label='95% CI')

# Mark monsoon shading
for year in [2021, 2022, 2023, 2024, 2025]:
    ax.axvspan(pd.Timestamp(f'{year}-06-01'), pd.Timestamp(f'{year}-09-30'),
                alpha=0.08, color='#3b82f6', label='Monsoon' if year == 2021 else '')

ax.set_title('🔮 ARIMA Forecast — Dengue Cases (Next 24 Weeks)',
             color='white', fontsize=14, fontweight='bold')
ax.set_xlabel('Date', color='#94a3b8')
ax.set_ylabel('Weekly Cases', color='#94a3b8')
ax.tick_params(colors='#94a3b8')
ax.legend(facecolor='#334155', edgecolor='none', labelcolor='white', fontsize=9)

plt.tight_layout()
plt.savefig('arima_forecast.png', dpi=130, bbox_inches='tight', facecolor='#0f172a')
plt.show()

print(f'✅ ARIMA(2,1,2)(1,1,1,52) model fitted')
print(f'   AIC: {arima_fit.aic:.2f} | BIC: {arima_fit.bic:.2f}')
print(f'   24-week forecast range: {forecast.values.min():.0f} – {forecast.values.max():.0f} cases/week')
```
